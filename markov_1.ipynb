First-order Markov chains exhibit the Markov property - the next state is only dependent on the current state, and not how it got there or the states prior.

s(t+1) = s(t).P

#### Calculating the stationary distribution
Usually referred to as pi.

```python
P = np.array([[0.2, 0.7, 0.1],
              [0.9, 0.0, 0.1],
              [0.2, 0.8, 0.0]
             ])
state = np.array([[1.0, 0.0, 0.0]])
state_hist = state
dfstate_hist = pd.DataFrame(state)
hist_distr = [[0, 0, 0]]
```

```python
for x in range(20):
    state = np.dot(state, P)
    print(state)
    state_hist = np.append(state_hist, state, axis=0)
    df_hist_distr = pd.DataFrame(state_hist)
    df_hist_distr.plot()
```
